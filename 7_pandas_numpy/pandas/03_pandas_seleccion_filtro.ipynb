```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "nombre":       ["Ana pérez", "Luis Lopez", "Eva Pérez", "Carlos Rodriguez", "Marta de Vargas", "Pedro Porro" ],
    "departamento": ["IT", "Ventas", "IT", "RRHH", "Ventas", "IT"],
    "edad":         [23, 31, 28, 45, 22, 38],
    "salario":      [28000, 42000, 35000, 61000, 24000, 55000]
})
```

# Selección y filtrado de datos con Pandas

Seleccionar filas y columnas específicas es la operación más frecuente en Pandas. Hay tres mecanismos principales: acceso directo por nombre, `loc` (por etiqueta) e `iloc` (por posición entera). Entender cuándo usar cada uno evita errores habituales.

```python
print(df[['nombre','edad']]) #para sacar varias series
```


## Loc - seleccion por etiqueta

loc['filas, columnas']. usa los nombres del indice y de la columnas. **A diferencia de slicing de python** si tiene los limite incluidos



```python
# fila por indice

print(df.loc[0]) # como devolver un diccionario el indice 0 de mi lista

print('-' * 30)

print(df.loc[1:3]) # OJO el tres esta incluido
```

El siguiente punto interesante sería juntar la posibilidad de hacer una combinación de ambos. Es dicer filtrar nombre y salario de el registro 3 al 5. 

```python
### Ejemplo

notas = pd.Series([1,2.5,3,4,5,6], 
                  index=['Ana', 'Maria', 'Pepe', 'Joaquin', 'Juan', 'Miguel'],
                  name = 'nota_final'
                  )
print(notas)
```

```python
print(df.loc[3:5, ['nombre', 'salario']])
```

## set_index() y idx_loc - recogemos el valor indices no numericos

```python

# crea un nuevo dataframe donde los indices sea al serie de nombre
df_idx = df.set_index('nombre')
#print(df_idx)
print(df_idx.loc['Eva'])
```

# Loc y df tiene un problema respecto al slicing que no tiene posiciones negativas.
ILOC que se basa en en slicing de python.

```python
print(df.iloc[-2:])
# me permite trabajar como en python


# iloc me permite trabajar con filas y columnas [filas, columnas ]
print(df.iloc[0:3, 0:2])
```

| | `loc` | `iloc` |
|-|-------|--------|
| Basado en | Etiquetas (nombres) | Posiciones enteras |
| Extremo final | **Incluido** | **Excluido** |
| Uso ideal | Índice significativo | Índice numérico o acceso posicional |

## Filtro con condiciones booleanas

Por ejemplo sacar los trabajadores que tenga un salario mayor que 30000. O un salario incluido en un rango 
  

```python

## and en dataframe es &. y or en dataframe es | 
rango = (df['salario'] >= 20000) & (df['salario'] <= 40000)
print(df[rango])
```

## Filtro con condiciones booleanas

Por ejemplo sacar los trabajadores que tenga un salario mayor que 30000. O un salario incluido en un rango 
  
  

```python

  # lo primero seria la condicion, es decir filtrar el df por la condicción

resultado = df['salario'] > 30000
print(df[resultado])

```

```python
## and en dataframe es &. y or en dataframe es | 
rango = (df['salario'] >= 20000) & (df['salario'] <= 40000)
print(df[rango])
```

## rango inclusivo

```python
rango_edades = df['edad'].between(20,30)
print(df[rango_edades])
```

## QUERY() - filtrado con expresión de texto tipo sql

Alternativa más legible para filtros complejos, inspirada en SQL

```python
# Equivale a df[(df["salario"] > 40000) & (df["departamento"] != "RRHH")]

resultado = df.query("salario > 40000 and departamento != 'RRHH'" )
print(resultado)

limite = 50000
resultado2 = df.query("salario >= @limite")
print(resultado2)

# buscar trabajadores con apellidos Perez
apellidos = df.query("nombre.str.contains('Pérez', case=False)")
print(apellidos)
```

**Resumen:**
- `df["col"]` → Series de una columna
- `df[["col1", "col2"]]` → DataFrame de varias columnas
- `df.loc[fila, col]` → por etiqueta (extremo incluido)
- `df.iloc[fila, col]` → por posición entera (extremo excluido)
- `df[condición]` → filtro booleano
- `df.query("expr")` → filtro en texto estilo SQL
