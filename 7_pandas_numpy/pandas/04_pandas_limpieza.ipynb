```python
import pandas as pd
```

```python
# Dataset con problemas típicos
df = pd.DataFrame({
    "ID":           [1, 2, 3, 4, 4, 5, 6],
    "Nombre":       ["Ana", "Luis", "Eva", "Carlos", "Carlos", None, "Marta"],
    "Departamento": ["IT", "ventas", "IT", "RRHH", "RRHH", "Ventas", "IT "],  # inconsistencias
    "Salario":      [28000, None, 35000, 61000, 61000, 48000, 24000],
    "Edad":         [23, 31, None, 45, 45, 38, 22],
})
print(df)
```

# Limpieza de datos

En proyecto reales, rara vez, los datos llegan de forma correcta o perfectas condiciones. Esta es una de las fases más importantes del trabajo, detectar y tratar los datos inconsistentes, valores nulos, duplicados, errores, etc

## Valores nulos: isnull, dropna, fillna.

```python

# Detectar nulos
print('Mapa de nulos')
print(df.isnull())

# quiero los nulos pero por columna
print(df.isnull().sum())
```

```python
# elimina todos los registros que tenga nulos
df_sin_nulos = df.dropna()
print(df_sin_nulos)

# eliminar solo algunas columnas especificas
df_sin_nulos2 = df.dropna(subset=['Salario', 'Edad'])
print(df_sin_nulos2)
```

```python
## fillna: rellenar nulos con un valor concreto 
df_relleno = df.copy()

# Valor fijo
df_relleno['Nombre'] = df_relleno['Nombre'].fillna('Desconocido')

print(df_relleno)
```

```python
#Sustituir nulos por mediana en salarios
mediana_salario = df_relleno['Salario'].median()
df_relleno['Salario'] = df_relleno['Salario'].fillna(mediana_salario)
```

```python
#Sustituir nulos por media en edad
media_edad = df_relleno['Edad'].mean()
df_relleno['Edad'] = df_relleno['Edad'].fillna(media_edad)

print(df_relleno)
```

## DUPLICADOS: duplicated(), drop_duplicates

```python
# detectar filas duplicadas.
print('Es duplicada??')
duplicado = df_relleno.duplicated()

#print(df[duplicado])

# eliminar duplicados (manteniendo la primera concurrencia por defecto)
df_resultante = df_relleno.drop_duplicates()
print(df_resultante)
```

## Renombrar columnas e indices del Dataframe

```python

df2 = df.copy()
df2.columns = df2.columns.str.lower()
print(df2)

print('-'*30)
# quiero cambiar todas las columnas por los nombre que a mi me vengan bien para tratar los datos.

df_rename = df_resultante.rename(columns={
    "ID": 'id',
    "Nombre": 'name',
    "Departamento": 'dpto',
    "Salario": 'salary',
    "Edad": 'age'
})

print(df_rename)
```

```python
# Convertir una serie o columna de un tipo a otro. de float a int
```

```python

df_rename['age'] = df_rename['age'].astype('Int64')
print(df_rename)
```

## Replace: reemplazar valores dentro de una serie

```python
# reemplazar Ventas => VENTAS

# primero limpiar espacios y llevar a mayusculas
#df_rename['dpto'] = df_rename['dpto'].str.strip().str.upper() ## VENTAS

# segunda forma de hacerlo, seria con replace
diccionario = {
    "Ventas": "COMERCIAL",
    "ventas": "COMERCIAL",
    "rrhh": "RRHH",
    "it": "IT"
}

df_rename['dpto'] = df_rename['dpto'].replace(diccionario) 

print(df_rename)
```

**Resumen — pipeline de limpieza típico:**
1. `df.isnull().sum()` → detectar nulos
2. `df.dropna()` o `df.fillna()` → tratar nulos
3. `df.duplicated()` → detectar duplicados
4. `df.drop_duplicates()` → eliminarlos
5. `df.rename()` / `.columns.str.lower()` → normalizar nombres
6. `.str.strip()` / `.replace()` → normalizar valores
