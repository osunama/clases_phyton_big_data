```python
import pandas as pd
```

```python

df = pd.DataFrame({
    "nombre":      ["Ana", "Luis", "Eva", "Carlos", "Marta", "Pedro", "Laura", "Tomás"],
    "departamento":["IT", "Ventas", "IT", "RRHH", "Ventas", "IT", "RRHH", "Ventas"],
    "edad":        [23, 31, 28, None, 22, 38, 35, 27],
    "salario":     [None, 42000, 35000, 61000, 24000, 55000, 48000, 31000],
    "activo":      [True, True, False, True, True, True, False, True],
    "fecha_alta":  pd.to_datetime(["2022-01-10", "2019-06-15", "2021-03-20",
                                   "2015-11-01", "2023-02-28", "2017-08-12",
                                   "2020-04-05", "2021-09-18"])
})
print(df)
```

## Exploración de columnas por categoría: value_counts, uninque, nunique

```python
# value_count: repeticion de cada valor 
print(df['departamento'].value_counts())

print(df['departamento'].value_counts(normalize=True) * 100) # a nivel visual mejor verlo en porcentaje pero a nivel de dato mejor sin el *100
```

```python
# lista de valores unicos por departamento
print (df['departamento'].unique())

#nunique: numero de valores unicos por columna
print (df.nunique())
```

## Pandas Detecta valeres nulos de manera sencilla

```python
#Tanto isnul() como isna() hace lo mismo
print(df.isnull())
print(df.isna())

# Total nulos
print(df.isnull().sum())

# total nulos departamento 
print(df['salario'].isnull().sum())
```

**Resumen — checklist de exploración inicial:**
1. `df.shape` → dimensiones
2. `df.head()` / `df.tail()` → ver los datos
3. `df.info()` → tipos y nulos de un vistazo
4. `df.describe()` → estadísticas numéricas
5. `df[col].value_counts()` → distribución de categóricas
6. `df.isnull().sum()` → conteo de nulos
