# Ejercicio: Análisis de Datos del Bootcamp
Contexto:
Tus alumnos acaban de recibir métricas sobre el rendimiento, práctica y satisfacción en un curso de desarrollo. Deben visualizar esta información usando tres tipos de gráficos diferentes.

```python

import matplotlib.pyplot as plt
import pandas as pd

# Datos 1: Evolución temporal
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
matriculas = [15, 22, 20, 28, 35, 40]

# Datos 2 y 3: Rendimiento por tecnología
datos_tech = {
    "Tecnologia": ["Python", "SQL", "Node.js", "React", "Angular"],
    "Horas_Practica_Semanal": [12, 6, 8, 10, 5],
    "Satisfaccion_Alumnos": [4.8, 4.1, 4.4, 4.7, 3.9]
}

```

## Misión:

Gráfico de Líneas (Evolución):
Muestra el crecimiento de las matriculas a lo largo de los meses.
Requisitos: La línea debe ser punteada (linestyle='--'), de color azul y tener un marcador circular (marker='o') en cada punto.

Gráfico de Barras Verticales (Carga de trabajo):
Compara las Horas_Practica_Semanal requeridas para cada Tecnologia.
Requisitos: Todas las barras deben ser de color naranja con un borde negro (edgecolor='black') y un grosor de línea de 1.5.

Gráfico de Barras Horizontales (Valoración):
Muestra la Satisfaccion_Alumnos obtenida por cada Tecnologia.
Requisitos: Pon un color distinto para cada barra,



```python
# Gráfico de Líneas (Evolución):
# Muestra el crecimiento de las matriculas a lo largo de los meses.
# Requisitos: La línea debe ser punteada (linestyle='--'), de color azul y tener un marcador circular (marker='o') en cada punto.

plt.figure(figsize=(10,6))
plt.plot(meses, matriculas,
         linestyle='--',
         color='blue',
         marker="o")
plt.xlabel('Meses')
plt.ylabel('Matriculas')
plt.title('Tendencias matrículas (Enero - Junio)')
plt.grid()
plt.show()
```

```python
# Gráfico de Barras Verticales (Carga de trabajo):
# Compara las Horas_Practica_Semanal requeridas para cada Tecnologia.
# Requisitos: Todas las barras deben ser de color naranja con un borde negro (edgecolor='black') y un grosor de línea de 1.5.
df_tech = pd.DataFrame(datos_tech)

plt.figure(figsize=(10,6))
plt.bar(
    df_tech['Horas_Practica_Semanal'],
    df_tech['Tecnologia'],
    color='orange',
    edgecolor='black',
    linewidth='1.5'

)
plt.title('Satisfacción lenguajes de programación')
plt.xlabel('Satisfación')
plt.ylabel('Lenguaje')
plt.show()
```

```python
# Gráfico de Barras Horizontales (Valoración):
# Muestra la Satisfaccion_Alumnos obtenida por cada Tecnologia.
# Requisitos: Pon un color distinto para cada barra

df_tech = pd.DataFrame(datos_tech)
colores = ['red', 'blue','green', 'gray', 'yellow']
plt.figure(figsize=(10,6))
plt.barh(
    df_tech['Tecnologia'],
    df_tech['Satisfaccion_Alumnos'],
    color=colores,
    edgecolor='black',
    linewidth='1.5'

)
plt.title('Satisfacción Alumnos')
plt.xlabel('Satisfación')
plt.ylabel('Lenguaje')
plt.show()
```
